**How many different year value are there**

```python
salaries.Year.value_counts()
```

**Print out max and avg Totalpay**

```python
#max
salaries.TotalPay.max()
```

```python
#avg
salaries.TotalPay.mean()
```

**How many records are there which have TotalPay bigger than avg TotalPay**

```python
len(salaries.TotalPay[salaries.TotalPay > salaries.TotalPay.mean()].value_counts())
# df.TotalPay[df.TotalPay>df.TotalPay.mean()].count()
```

**How much does ALBERT PARDINI make (including benefits)?**

```python
salaries.TotalPayBenefits[salaries["EmployeeName"]=="ALBERT PARDINI"]
```

**Who is lowest paid person**

```python
salaries[salaries.TotalPayBenefits==salaries.TotalPayBenefits.min()]["EmployeeName"]
```

**What are the unique Jobs in dataset?**

```python
salaries.JobTitle.unique()
```

**List 3 least common Jobs**

```python
salaries["JobTitle"].value_counts().tail(3)
```

**How many people earned more than avg in 2012**

```python
salaries[(salaries.TotalPayBenefits>salaries.TotalPayBenefits.mean()) & (salaries.Year==2012)]["EmployeeName"].count()
```

**How many people have the word Manager in their job title?**

```python
def manager_func(x):
    if 'manager' in x.lower():
        return True
    else:
        return False
```

```python
len(salaries[salaries.JobTitle.apply(lambda x:manager_func(x))])
```
